# Data Cleaning

### Date Time

```python
df = df.sort_values(by='Date')
df
```

### Null Values

Dropped the one row with null values in it  

```python
df = df[df['Div'] == 'E0']
```

```python
df.isna().sum()
```

### Dropping, Renaming, and Creating Columns 

First, I'm going to drop betting data that I won't be using in my model

```python
# Dropping irrevelant betting columns 
# Focusing mainly on 365 bets
# Should clear the rest of the null values
df.drop(df.iloc[:, 26:], inplace=True, axis=1)
```

```python
df
```

```python
df.isna().sum()
```

Checking the rest of the columns, and deciding which ones to drop

```python
df.columns.values
```

```python
# Dropped data to only show columns that I would only know before a match 
df.drop(['Div', 'Referee', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST', 'HF',
       'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR',  ], axis=1, inplace=True)
```

```python
df.head()
```

Renaming columns for better understanding of dataset

```python
df.columns = ['Date','HomeTeam','AwayTeam','FT_Home_Goal','FT_Away_Goal','FT_Result', 
              'HomeOdds', 'DrawOdds', 'AwayOdds']
```

Created a winners column to see if the home or away team won, or they drew

```python
# Function 1 
def Winner(Result):
    '''
    This function checks if the results at the end were either the home team, away team, or they drew. 
    Then creating a new column with 1 being home team, -1 away team, and 0 they drew.
    '''
    if Result == 'H':
        return 1
    if Result == 'A':
        return -1
    else:
        return 0
# create a new column based on condition
df['Winner'] = df['FT_Result'].apply(Winner)
# display the dataframe
df
```

### Merging Both Datasets Into One

Since the first dataset begins midway through the 2003 season. I'm going to make it start with the 2004 season. The first dataset goes all the way to the 2020 season, but the second dataset ends in 2018. Thus, I'm going to focus on the seasons from 2004-2018
