# Data Exploration 

The Premier League data for each season from 2003 - 2020 were all in individual CSV files. Therefore, I ran an os.path.join() to join the paths into one. Then created a list using glob. that looks for specific patterns.

```python
# merging the files
joined_files = os.path.join("data/EPL/", "EPL*.csv")
  
# # A list of all joined files is returned
joined_list = glob.glob(joined_files)
```

```python
joined_list
```

Created a for loop to check on every CSV, so there wouldn't be an issue when concat later

```python
for x in range(len(joined_list)):
    print(joined_list[x])
    test = pd.read_csv(joined_list[x])
    
    print(test.shape)
    print(test.columns)
    print("*"*20)
```

```python
# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
df
```
